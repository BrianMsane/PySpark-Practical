## üìÇ Uploading Files in Colab

## üîê Environment Variables & Secrets in Google Colab

When working in Colab, you‚Äôll often need to use **API keys, tokens, or credentials**.  
There are three main ways to handle them:

1. **Environment variables (`os.environ`)** ‚Üí quick & temporary.  
2. **`.env` files with `python-dotenv`** ‚Üí reloadable, can be stored in Drive.  
3. **Colab Secrets Manager** ‚Üí more secure, hidden values tied to the notebook (Pro/Enterprise).  


**1. Environment Variables with os.environ**

```python
import os

# Set a variable (only lasts for this runtime session)
os.environ["MY_API_KEY"] = "12345-ABCDE"

# Get the variable
print("API Key:", os.environ.get("MY_API_KEY"))
```

‚ö†Ô∏è These variables are not persistent ‚Äî they disappear when the runtime resets.

**2. Using .env Files with python-dotenv**

```python
# Install dotenv
!pip install python-dotenv

import os
from dotenv import load_dotenv

# Create a .env file
with open(".env", "w") as f:
    f.write("API_KEY=abcd1234\nUSERNAME=myuser\n")

# Load from .env file
load_dotenv(".env")

# Access values
print("API_KEY:", os.getenv("API_KEY"))
print("USERNAME:", os.getenv("USERNAME"))
```

üí° Store your .env in Google Drive so it persists across sessions.

**3. Using Colab Secrets**

#### üîë Colab Secrets

Colab provides a **Secrets manager** that stores values securely.

- Open the left sidebar ‚Üí **üîë Secrets tab**.  
- Add your secret (e.g., `OPENAI_API_KEY`).  
- Secrets are encrypted and not visible in the notebook.  
- You can access them using `os.environ` just like normal variables.


```python
import os

# If you saved a secret named OPENAI_API_KEY in Colab's UI:
api_key = os.environ.get("OPENAI_API_KEY")
print("Loaded API key from Colab Secrets:", api_key)
```

Secrets Best Practices
- ‚úÖ Use **Colab Secrets** for sensitive info (best option).
- ‚úÖ Use `.env` files stored in Google Drive for reusability.
- ‚ùå Avoid hardcoding secrets directly in notebooks.
- ‚ùå Don‚Äôt share notebooks that print or expose secrets.
